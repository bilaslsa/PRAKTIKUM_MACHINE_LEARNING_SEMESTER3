```python
# Memanggil dan Membaca Dataset
import pandas as pd
df = pd.read_csv('/content/gdrive/MyDrive/Praktikum Mandiri 2/Praktikum 02/data/day.csv')
df
```

```python
# -- Menghitung Pembagian Data --

#Menghitung total data keseluruhan
jumlah_data = len(df)

# Menentukan jumlah data testing (20% dari total)
jmlh_data_testing = int(0.2 * jumlah_data)

# Sisa data untuk training dan testing 80% dari total data
jmlh_training_testing = jumlah_data - jmlh_data_testing

# Menentukan jumlah data validasi (10% dari sisa data)
jmlh_data_validation = int (0.1 * jmlh_training_testing)

# Jumlah data training sisa dari semua data
jmlh_data_training = jmlh_training_testing - jmlh_data_validation

print(f"Jumlah data training adalah: {jmlh_data_training}")
print(f"Jumlah data validation adalah: {jmlh_data_validation}")
print(f"Jumlah data testing adalah: {jmlh_data_testing}")
print(f"Total seluruh data adalah: {jmlh_data_training + jmlh_data_validation + jmlh_data_testing}")
```

```python
# Menampilkan 5 baris pertama data training
display(df[:jmlh_data_training].head())

# Menampilkan 5 baris pertama dari data validation
print("\n========================================")
print("         (b) DATA VALIDATION")
print("========================================")
print("Lima baris teratas dari data validation:")
display(df[jmlh_data_training:jmlh_data_training+jmlh_data_validation].head())

# Menampilkan 5 baris pertama dari data testing
print("\n========================================")
print("          (c) DATA TESTING")
print("========================================")
print("Lima baris teratas dari data testing:")
display(df[jmlh_data_training+jmlh_data_validation:].head())
```
